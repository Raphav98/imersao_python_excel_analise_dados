```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/acoes_pura.xlsx",sheet_name="Principal")
df_principal
```

```python
df_Total_de_acoes = pd.read_excel("/content/acoes_pura.xlsx",sheet_name="Total_de_acoes")
df_Total_de_acoes
```

```python
df_Ticket = pd.read_excel("/content/acoes_pura.xlsx",sheet_name="Ticket")
df_Ticket
```

```python
df_Tabela_segm_idd = pd.read_excel("/content/acoes_pura.xlsx",sheet_name="Tabela_segm_idd")
df_Tabela_segm_idd
```

```python
df_principal = df_principal[['Ativo',	'Data','Último (R$)','Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['Var_inicial'] = df_principal['valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_Total_de_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Varicao_rs'] = (df_principal['valor_final'] - df_principal['Var_inicial'] )* df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Varicao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_Ticket, left_on='Ativo', right_on='Ticker', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_Tabela_segm_idd, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade em Anos':'Idade'}).copy()
df_principal
```

```python
df_principal['Cat_Idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100 anos' if x > 100 else 'Menos de 50' if x < 50 else 'Entre 50 e 100 anos')
df_principal
```

```python
# Calcular estatísticas
maior = df_principal['Varicao_rs'].max()
menor = df_principal['Varicao_rs'].min()
media_geral = df_principal['Varicao_rs'].mean()
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Varicao_rs'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Varicao_rs'].mean()

# Imprimir as estatísticas formatadas
print("Maior:", "R$ {:,.2f}".format(maior))
print("Menor:", "R$ {:,.2f}".format(menor))
print("Média:", "R$ {:,.2f}".format(media_geral))
print("Média de quem subiu:", "R$ {:,.2f}".format(media_subiu))
print("Média de quem desceu:", "R$ {:,.2f}".format(media_desceu))
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Varicao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Varicao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y= 'Varicao_rs', text='Varicao_rs', title='Variação Reais por Resultado')
fig.update_traces(texttemplate='R$ %{text:.2f}', textposition='outside')
fig.show()
```
